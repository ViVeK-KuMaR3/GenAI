```python
from langchain.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Pinecone
import pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
from langchain import HuggingFacePipeline
from huggingface_hub import notebook_login
import torch
```

```python
URLs = [
    "https://ineuron.ai/"
]
```

```python
loader=UnstructuredURLLoader(urls=URLs)
```

```python
data=loader.load()
```

```python
data
```

```python
splitter=CharacterTextSplitter(separator='\n',chunk_size=1000,chunk_overlap=200)
```

```python
text_chunks=splitter.split_documents(data)
```

```python
text_chunks
```

```python
embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
query_result = embeddings.embed_query("Hello world")
len(query_result)
```

```python
import os

#PINECONE_API_KEY = os.environ.get('PINECONE_API_KEY', '6ee7dc8a-43fb-4480-8774-f1d02577386d')
PINECONE_API_KEY=os.environ.get('PINECONE_API_KEY', '81bd462d-e93c-4547-be15-d827405c4d13')
PINECONE_API_ENV = os.environ.get('PINECONE_API_ENV', 'gcp-starter')
```

```python
import pinecone
# initialize pinecone
pinecone.init(
    api_key=PINECONE_API_KEY,  # find at app.pinecone.io
    environment=PINECONE_API_ENV  # next to api key in console
)
index_name = "chatbot" # put in the name of your pinecone index here
```

```python
docsearch = Pinecone.from_texts([t.page_content for t in text_chunks], embeddings, index_name=index_name)
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True,)

model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
```

```python
pipe = transformers.pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.predict("Please provide a concise summary of the Book Harry Potter")
```

```python
from langchain.chains import RetrievalQA
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=docsearch.as_retriever())
```

```python
query='Tell me the course price of Full Stack Data Science with Generative AI provide by ineuron'
print(qa.run(query))
```
